```
import pandas as pd
```

```
#load the data
df = pd.read_csv('/dbfs/mnt/candev-data/Workshop/Tabular/csbcourtfinderlistingforbcoct2016.csv') # available from https://open.canada.ca/data/en/dataset/c95a2ad5-f62a-43d6-8678-80a617b6200e

print('Number of records: %d' % df.shape[0])
print('Number of columns: %d' % df.shape[1])

print('\nColumn names: %s' % df.columns)
```

```
#check the first five records
df.head()
```

```
#take a subset of the columns
df_2 = df.loc[:, ['Placemark_name', 'Latitude', 'Longitude', 'Court_Level']]

print('Number of records: %d' % df_2.shape[0])
print('Number of columns: %d' % df_2.shape[1])

print('\nColumn names: %s' % df_2.columns)
```

```
#check the first five records
df_2.head()
```

```
#take a subset of the rows
df_3 = df_2.loc[20:30, :]

print('Number of records: %d' % df_3.shape[0])
print('Number of columns: %d' % df_3.shape[1])

print('\nColumn names: %s' % df_3.columns)
```

```
#check the first five records
df_3.head()
```

```
#select only records with a latitude between 49 to 50 and longitude between -124 to -123
df_4 = df_2[(df_2['Latitude'] > 49) & (df_2['Latitude'] < 50) & (df_2['Longitude'] > -124) & (df_2['Longitude'] < -123)]

print('Number of records: %d' % df_4.shape[0])
print('Number of columns: %d' % df_4.shape[1])

print('\nColumn names: %s' % df_4.columns)
```

```
#check the first five records
df_4.head()
```

```
#check the different court levels
print('Court Levels:')
print(df_4['Court_Level'].unique())
```

```
#select only rows with a court level of Supreme/Appeal
df_5 = df_4[df_4['Court_Level'] == 'Supreme/Appeal']

print('Number of records: %d' % df_5.shape[0])
print('Number of columns: %d' % df_5.shape[1])

print('\nColumn names: %s' % df_5.columns)
```
